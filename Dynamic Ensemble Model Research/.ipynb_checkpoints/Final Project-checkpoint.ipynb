```python
x=np.array(df.iloc[:,:-1])
y=np.array(df.quality)
x.shape,y.shape
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=10)
xtrain.shape,ytrain.shape,xtest.shape,ytest.shape
```

```python
def weightInitialization(n_features):
    w = np.zeros((1,n_features))
    b = 0
    return w,b

def sigmoid_activation(result):
    final_result = 1/(1+np.exp(-result))
    return final_result

def model_optimize(w, b, X, Y):
    m = X.shape[0]
    
    #Prediction
    final_result = sigmoid_activation(np.dot(w,X.T)+b)
    Y_T = Y.T
    cost = (-1/m)*(np.sum((Y_T*np.log(final_result)) + ((1-Y_T)*(np.log(1-final_result)))))
    #
    
    #Gradient calculation
    dw = (1/m)*(np.dot(X.T, (final_result-Y.T).T))
    db = (1/m)*(np.sum(final_result-Y.T))
    
    grads = {"dw": dw, "db": db}
    
    return grads, cost

def model_predict(w, b, X, Y, learning_rate, no_iterations):
    costs = []
    for i in range(no_iterations):
        #
        grads, cost = model_optimize(w,b,X,Y)
        #
        dw = grads["dw"]
        db = grads["db"]
        #weight update
        w = w - (learning_rate * (dw.T))
        b = b - (learning_rate * db)
        #
        
        if (i % 100 == 0):
            costs.append(cost)
            #print("Cost after %i iteration is %f" %(i, cost))
    
    #final parameters
    coeff = {"w": w, "b": b}
    gradient = {"dw": dw, "db": db}
    
    return coeff, gradient, costs

def predict(final_pred, m):
    y_pred = np.zeros((1,m))
    for i in range(final_pred.shape[1]):
        if final_pred[0][i] > 0.5:
            y_pred[0][i] = 1
    return y_pred
```

```python
#Get number of features
n_features = xtrain.shape[1]
print('Number of Features', n_features)
w, b = weightInitialization(n_features)
#Gradient Descent
coeff, gradient, costs = model_predict(w, b, xtrain, ytrain, learning_rate=0.0001,no_iterations=10000)
#Final prediction
w = coeff["w"]
b = coeff["b"]
print('Optimized weights', w)
print('Optimized intercept',b)
#
final_train_pred = sigmoid_activation(np.dot(w,xtrain.T)+b)
final_test_pred = sigmoid_activation(np.dot(w,xtest.T)+b)
#
m_tr =  xtrain.shape[0]
m_ts =  xtest.shape[0]
#
y_tr_pred = predict(final_train_pred, m_tr)
print('Training Accuracy',accuracy_score(y_tr_pred.T, ytrain))
#
y_ts_pred = predict(final_test_pred, m_ts)
print('Test Accuracy',accuracy_score(y_ts_pred.T, ytest))
```

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

class LogisticRegression():

    def __init__(self, lr=0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)

            dw = (1/n_samples) * np.dot(X.T, (predictions - y))
            db = (1/n_samples) * np.sum(predictions-y)

            self.weights = self.weights - self.lr*dw
            self.bias = self.bias - self.lr*db


    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y<=0.5 else 1 for y in y_pred]
        return class_pred
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class EnsembleLogisticRegression:
    def __init__(self, lr_middle=0.1, lr_left=0.1, lr_right=0.1, n_iters=1000):
        self.lr_middle = lr_middle
        self.lr_left = lr_left
        self.lr_right = lr_right
        self.n_iters = n_iters
        self.theta_M = None
        self.theta_L = None
        self.theta_R = None

    def fit(self, X, y):
        # Initialize parameters
        n_samples, n_features = X.shape
        self.theta_M = np.zeros(n_features)
        self.theta_L = np.zeros(n_features)
        self.theta_R = np.zeros(n_features)

        # Training LR_middle
        for _ in range(self.n_iters):
            linear_pred_M = np.dot(X, self.theta_M)
            predictions_M = sigmoid(linear_pred_M)
            
            # Training LR_left
            linear_pred_L = np.dot(X, self.theta_L)
            predictions_L = sigmoid(linear_pred_L)
           
            # Training LR_right
            linear_pred_R = np.dot(X, self.theta_R)
            predictions_R = sigmoid(linear_pred_R)

            #Optimizing Weights
            dw_M = (1 / n_samples) * np.dot(X.T, (predictions_M - y))
            self.theta_M -= self.lr_middle * dw_M
            dw_L = (1 / n_samples) * np.dot(X.T, (predictions_L - y))
            self.theta_L -= self.lr_left * dw_L
            dw_R = (1 / n_samples) * np.dot(X.T, (predictions_R - y))
            self.theta_R -= self.lr_right * dw_R

        

    def predict_proba(self, X):
        linear_pred_M = np.dot(X, self.theta_M)
        predictions_M = sigmoid(linear_pred_M)

        linear_pred_L = np.dot(X, self.theta_L)
        predictions_L = sigmoid(linear_pred_L)

        linear_pred_R = np.dot(X, self.theta_R)
        predictions_R = sigmoid(linear_pred_R)

        # Calculate overall probability
        prob_left = (predictions_L * predictions_M) + (predictions_R * (1 - predictions_M))
        prob_right = ((1-predictions_L)*predictions_M)+(1-predictions_R)*(1-predictions_M)

        return np.column_stack((prob_left, prob_right))

    def predict(self, X, threshold=0.5):
        proba = self.predict_proba(X)
        return np.where(proba[:, 0] >= threshold, 1, 0)


model = EnsembleLogisticRegression()
model.fit(xtrain, ytrain)

# Make predictions on the test set
predictions = model.predict(xtest)

accuracy_score(predictions,ytest)
```

```python
class ThreeLayerEnsembleLogisticRegression:
    def __init__(self, lr_middle=0.01, lr_left=0.01, lr_right=0.01, n_iters=1000):
        self.lr_middle = lr_middle
        self.lr_left = lr_left
        self.lr_right = lr_right
        self.n_iters = n_iters
        self.theta_M = None
        self.theta_L = None
        self.theta_R = None
        self.theta_LL = None  # LR_left's LR
        self.theta_LR = None  # LR_left's RR
        self.theta_RL = None  # LR_right's LR
        self.theta_RR = None  # LR_right's RR

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize parameters
        self.theta_M = np.random.rand(n_features)
        self.theta_L = np.random.rand(n_features)
        self.theta_R = np.random.rand(n_features)
        self.theta_LL = np.random.rand(n_features)
        self.theta_LR = np.random.rand(n_features)
        self.theta_RL = np.random.rand(n_features)
        self.theta_RR = np.random.rand(n_features)
        # self.theta_M =np.zeros(n_features)
        # self.theta_L = np.zeros(n_features)
        # self.theta_R = np.zeros(n_features)
        # self.theta_LL = np.zeros(n_features)
        # self.theta_LR = np.zeros(n_features)
        # self.theta_RL = np.zeros(n_features)
        # self.theta_RR = np.zeros(n_features)

        for _ in range(self.n_iters):
            # First layer LR_middle
            linear_pred_M = np.dot(X, self.theta_M)
            predictions_M = self.sigmoid(linear_pred_M)

            # Second layer LR_left
            linear_pred_L = np.dot(X, self.theta_L)
            predictions_L = self.sigmoid(linear_pred_L)

            # Second layer LR_right
            linear_pred_R = np.dot(X, self.theta_R)
            predictions_R = self.sigmoid(linear_pred_R)

            # Third layer LR_left's LR
            linear_pred_LL = np.dot(X, self.theta_LL)
            predictions_LL = self.sigmoid(linear_pred_LL)

            # Third layer LR_left's RR
            linear_pred_LR = np.dot(X, self.theta_LR)
            predictions_LR = self.sigmoid(linear_pred_LR)

            # Third layer LR_right's LR
            linear_pred_RL = np.dot(X, self.theta_RL)
            predictions_RL = self.sigmoid(linear_pred_RL)

            # Third layer LR_right's RR
            linear_pred_RR = np.dot(X, self.theta_RR)
            predictions_RR = self.sigmoid(linear_pred_RR)

            # Update weights
            dw_M = (1 / n_samples) * np.dot(X.T, (predictions_M - y))
            self.theta_M -= self.lr_middle/100 * dw_M

            dw_L = (1 / n_samples) * np.dot(X.T, (predictions_L - y))
            self.theta_L -= self.lr_left*10 * dw_L

            dw_R = (1 / n_samples) * np.dot(X.T, (predictions_R - y))
            self.theta_R -= self.lr_right*100 * dw_R

            dw_LL = (1 / n_samples) * np.dot(X.T, (predictions_LL - y))
            self.theta_LL -= self.lr_left*10 * dw_LL

            dw_LR = (1 / n_samples) * np.dot(X.T, (predictions_LR - y))
            self.theta_LR -= self.lr_right*100 * dw_LR

            dw_RL = (1 / n_samples) * np.dot(X.T, (predictions_RL - y))
            self.theta_RL -= self.lr_left*10 * dw_RL

            dw_RR = (1 / n_samples) * np.dot(X.T, (predictions_RR - y))
            self.theta_RR -= self.lr_right*100 * dw_RR

    def predict_proba(self, X):
        linear_pred_M = np.dot(X, self.theta_M)
        predictions_M = self.sigmoid(linear_pred_M)

        linear_pred_L = np.dot(X, self.theta_L)
        predictions_L = self.sigmoid(linear_pred_L)

        linear_pred_R = np.dot(X, self.theta_R)
        predictions_R = self.sigmoid(linear_pred_R)

        linear_pred_LL = np.dot(X, self.theta_LL)
        predictions_LL = self.sigmoid(linear_pred_LL)

        linear_pred_LR = np.dot(X, self.theta_LR)
        predictions_LR = self.sigmoid(linear_pred_LR)

        linear_pred_RL = np.dot(X, self.theta_RL)
        predictions_RL = self.sigmoid(linear_pred_RL)

        linear_pred_RR = np.dot(X, self.theta_RR)
        predictions_RR = self.sigmoid(linear_pred_RR)

        # Calculate overall probability
        prob_left = (predictions_LL * predictions_L * predictions_M) + (predictions_RL * predictions_R * (1 - predictions_M))
        prob_right = ((predictions_LR) * predictions_L * predictions_M) + (predictions_RR * (1 - predictions_R) * (1 - predictions_M))

        return np.column_stack((prob_left, prob_right))

    def predict(self, X, threshold=0.5):
        proba = self.predict_proba(X)
        return np.where(proba[:, 0] >= threshold, 1, 0)


model = ThreeLayerEnsembleLogisticRegression()
model.fit(xtrain, ytrain)

# Make predictions on the test set
predictions = model.predict(xtest)
pred =  model.predict(xtrain)
# You can also get the probabilities for each class
probabilities = model.predict_proba(xtest)
accuracy_score(predictions, ytest),accuracy_score(pred,ytrain)
```
